```python
import json
from glob import glob
import pandas as pd
import numpy as np
```

```python
outputs = glob("output/*.json")
results = pd.DataFrame()

for fname in outputs:
  with open(fname) as f:
    output = json.load(f)
  
  metrics = {k:sum([r["metrics"][k] for r in output["results"]]) / len(output["results"]) for k in output["results"][0]["metrics"].keys()}
  to_add = { 
    "fname": fname,
    "dataset": output["args"]["dataset"],
    "model": output["args"]["model"],
    "template": output["args"]["template"],
    "max_length": output["args"]["max_length"],
  **metrics }
  df = pd.DataFrame([to_add.values()], columns=list(to_add.keys()))

  results = pd.concat([results, df], ignore_index=True)
```
